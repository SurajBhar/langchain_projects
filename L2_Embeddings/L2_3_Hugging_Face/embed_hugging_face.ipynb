# Sentence Transformers on Hugging Face
- Hugging Face sentence transformers is a python framework for state of the art sentence , text and image embeddings.
- One of the ebedding models is used in the HuggingFaceEmbeddings class.
- We have also added an alias for SentenceTransformerEmbeddings for users who are familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text = " My name is Suraj Bhardwaj. This is a test document."
query_result = embeddings.embed_query(text)
print(query_result)
```

```python
len(query_result)
```

```python
text_1 = " My name is Suraj Bhardwaj. This is a test document."
text_2 = " My name is Suraj Bhardwaj. This is a test document."
query_result = embeddings.embed_documents([text_1, text_2])
print(query_result)
```
